```python
import pygal
from IPython.display import Image

# Create a radar chart object
dot_chart = pygal.Radar()

# Set the title of the chart
dot_chart.title = 'Total Precipitation (mm)'

# Set the labels for the x-axis
dot_chart.x_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Add data for each year
dot_chart.add('2012', [67.81, 25.41, 30.72, 78.49, 87.12, 119.62, 98.56, 78.23, 86.11, 66.53, 25.65, 149.59])
dot_chart.add('2013', [28.68, 120.89, 41.90, 34.80, 81.54, 266.70, 91.67, 46.73, 56.14, 15.48, 69.09, 117.35])
dot_chart.add('2014', [82.31, 96.01, 102.36, 82.30, 72.63, 66.55, 116.07, 44.44, 17.77, 148.06, 133.86, 166.61])

# Render the chart to a PNG file
dot_chart.render_to_png('radar.png')

# Display the image in Jupyter Notebook
Image('radar.png')
```

```python
import pygal
from IPython.display import Image

# Create a gauge chart object
gauge_chart = pygal.Gauge()

# Set the title of the chart
gauge_chart.title = 'Total Precipitation (mm)'

# Set the range for the gauge
gauge_chart.range = [0, 1000]

# Add data for each year
gauge_chart.add('2012', 765)
gauge_chart.add('2013', 860)
gauge_chart.add('2014', 962)

# Render the chart to a PNG file
gauge_chart.render_to_png('gauge.png')

# Display the image in Jupyter Notebook
Image('gauge.png')
```
